```python
#学習データ、テストデータの読み込み
train0 = pd.read_csv('train.csv')
test0 = pd.read_csv('test.csv')
```

```python
train0
```

```python
def ratio_1(df):
    ratio = df[df['MIS_Status'] == 1].shape[0] / df.shape[0]
    return ratio, df.shape[0]
```

```python
#lowdoc
train0['LowDoc'] = train0['LowDoc'].fillna('UNK')
df_LD_Y = train0[train0['LowDoc'] == 'Y']
df_LD_S = train0[train0['LowDoc'] == 'S']
df_LD_N = train0[train0['LowDoc'] == 'N']
df_LD_A = train0[train0['LowDoc'] == 'A']
df_LD_0 = train0[train0['LowDoc'] == '0']
df_LD_UNK = train0[train0['LowDoc'] == 'UNK']

print(ratio_1(df_LD_Y),ratio_1(df_LD_S),ratio_1(df_LD_N),ratio_1(df_LD_A),ratio_1(df_LD_0),ratio_1(df_LD_UNK))
```

```python
#RevLineCr
train0['RevLineCr'] = train0['RevLineCr'].fillna('UNK')
df_RLC_Y = train0[train0['RevLineCr'] == 'Y']
df_RLC_T = train0[train0['RevLineCr'] == 'T'] 
df_RLC_N = train0[train0['RevLineCr'] == 'N']
df_RLC_0 = train0[train0['RevLineCr'] == '0']
df_RLC_UNK = train0[train0['RevLineCr'] == 'UNK']

print(ratio_1(df_RLC_Y),ratio_1(df_RLC_T),ratio_1(df_RLC_N),ratio_1(df_RLC_0),ratio_1(df_RLC_UNK))
```

```python
train0['DisbursementDate'] = train0['DisbursementDate'].fillna('UNK')
df_DD_0 = train0[train0['DisbursementDate'] != 'UNK']
df_DD_UNK = train0[train0['DisbursementDate'] == 'UNK']

print(ratio_1(df_DD_0),ratio_1(df_DD_UNK))
```
